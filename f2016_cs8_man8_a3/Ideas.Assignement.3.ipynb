```python
masterFile = 'f2016_cs8_a3.data.txt'
```

```python
dataFiles = [file.rstrip('\n') for file in open(masterFile,'r')]
```

```python
dataFiles
```

```python
def getDataFromFile(file):
    # output dictionary
    output = []
    # number of lines read
    nlines = 0
    # read file
    for line in open(file,'r'):
        # update number of lines read
        nlines += 1
        # exclude first line
        if "distance" in line:
            continue
        temp1 = line.rstrip('\n').split(',')
        try:
            output.append({'name': temp1[0].strip(' '), 'distance':float(temp1[1])})
        except:
            print('Invalid row : '+line.rstrip('\n'))
    return output, nlines
```

```python
# read all the files and place the content of each one of them in an element of the list
data = [getDataFromFile(file) for file in dataFiles]
```

```python
sum([item[1] for item in data])
```

```python
# flattens the list => complete list
cl = [item2 for item1 in data for item2 in item1[0]]
```

```python
cl
```

```python
# total distance run by everybody
totalDistanceRun = sum([item['distance'] for item in cl])
```

```python
# complete list of names
clNames = [item['name'] for item in cl]
```

```python
# initialize
totalDistance = 0
totalDistancePerPerson = {}
personAppearances = {}
appearences = {}

for item in cl:
    totalDistance += item['distance']
    if not item['name'] in totalDistancePerPerson.keys():
        totalDistancePerPerson[item['name']] = 0
        personAppearances[item['name']] = 0
    totalDistancePerPerson[item['name']] += item['distance']
    personAppearances[item['name']] += 1
    
for person in personAppearances.keys():
    if not personAppearances[person] in appearences.keys():
        appearences[personAppearances[person]] = []
    appearences[personAppearances[person]].append(person)
    
```

```python
participantDistances = {}
for item in cl:
    if not item['name'] in participantDistances.keys():
        participantDistances[item['name']] = []
    participantDistances[item['name']].append(item['distance'])
    
participantDistanceRun = {key: sum(values) for key, values in participantDistances.items()}
```
