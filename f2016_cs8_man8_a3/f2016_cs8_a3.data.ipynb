```python
import numpy as np
import pandas as pd
```

```python
files = [ { 'input':'f2016_cs8_a3.data.1.orig.txt', 'output':'f2016_cs8_a3.data.1.csv'},
          { 'input':'f2016_cs8_a3.data.2.orig.txt', 'output':'f2016_cs8_a3.data.2.csv'},
          { 'input':'f2016_cs8_a3.data.3.orig.txt', 'output':'f2016_cs8_a3.data.3.csv'}]
```

```python
output = []
for io in files:
    print('Input file'+io['input'])
    # read file in data frame
    td = pd.read_csv(io['input'],header=None)
    # add distance column
    td['distance'] = np.random.uniform(0,100,len(td))
    # set columns names
    td.columns=['name','distance']
    # save to output file
    print('Output file '+io['output'])
    td.to_csv(io['output'],header=True,index=False)
    
```
